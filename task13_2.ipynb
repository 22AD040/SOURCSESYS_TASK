```python
import numpy as np
import pandas as pd

data = {
    "CustomerID": np.arange(1, 11),
    "Purchase": np.array([1200, 500, 700, 1500, 300, 800, 950, 400, 1100, 650]),
    "Age": np.array([25, 32, 40, 29, 22, 35, 41, 28, 33, 27])
}

df = pd.DataFrame(data)

df["Category"] = np.where(df["Purchase"] > 1000, "Premium", "Normal")

df["Discount"] = df["Purchase"] * 0.9

print(df)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["A", "B", "C", None, "E"],
    "Score": [90, np.nan, 80, 70, np.nan]
})

df["Score"] = df["Score"].fillna(df["Score"].mean())
df["Name"] = df["Name"].fillna("Unknown")

print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Department": ["AI", "AI", "HR", "HR", "Sales"],
    "Salary": [50000, 60000, 45000, 48000, 55000]
})

group = df.groupby("Department").mean()

print(group)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Text": ["AI is good", "Gen AI rocks", "Data Science"]
})

df["Length"] = df["Text"].apply(len)

print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Product": ["A", "B", "C", "D"],
    "Price": [100, 200, 300, 400]
})

expensive = df[df["Price"] > 200]

print(expensive)
```
