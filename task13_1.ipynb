```python
import numpy as np

temps = np.array([32, 33, 35, 31, 30, 29, 28, 33, 36, 34])

fahrenheit = (temps * 9/5) + 32

abnormal = temps[(temps < 30) | (temps > 35)]

print("Temps:", temps)
print("Fahrenheit:", fahrenheit)
print("Abnormal:", abnormal)
```

```python
import numpy as np

data = np.array([120, 130, 115, 140, 150, 110, 125])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Std Dev:", np.std(data))
print("Min:", np.min(data))
print("Max:", np.max(data))
```

```python
import numpy as np

A = np.array([[1, 2],
             [3, 4]])

B = np.array([[5, 6],
             [7, 8]])

result = np.dot(A, B)

det = np.linalg.det(A)

print("Matrix Multiply:\n", result)
print("Determinant:", det)
```

```python
import numpy as np

data = np.array([100, 200, 300, 400, 500])

normalized = (data - data.min()) / (data.max() - data.min())

print(normalized)
```
