```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

months = np.arange(1, 13)
sales = np.array([12000, 13500, 12800, 15000, 16000, 17500,
                 18000, 19000, 20000, 21000, 22000, 24000])

df = pd.DataFrame({"Month": months, "Sales": sales})

df["MovingAvg"] = df["Sales"].rolling(window=3).mean()

plt.figure()
plt.plot(df["Month"], df["Sales"])
plt.plot(df["Month"], df["MovingAvg"])
plt.title("Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
df["GrowthRate"] = df["Sales"].pct_change() * 100

plt.figure()
plt.bar(df["Month"], df["GrowthRate"])
plt.title("Growth Rate")
plt.xlabel("Month")
plt.ylabel("Growth %")
plt.show()
```
